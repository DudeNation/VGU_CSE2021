# Relational  Operators

```c++
#include <stdio.h>

// + operator
printf("1 + 2 = %d\n\n", 1+2);
```

```c++
#include <stdio.h>

// - operator
printf("1 - 2 = %d\n\n", 1-2);
```

```c++
#include <stdio.h>

// * operator
printf("1 * 2 = %d\n\n", 1*2);
```

```c++
#include <stdio.h>

// / operator
printf("1 / 2 = %d\n", 1/2);
printf("1 / 2 = %f\n\n", 1/2);
printf("1 / 2 = %f\n", (float)1/2);
printf("1 / 2 = %f\n", 1/(float)2);
printf("1 / 2 = %f\n", (float)1/(float)2);
printf("1 / 2 = %f\n\n", (float)(1/2));
printf("1 / 2 = %f\n", 1.0/2);
printf("1 / 2 = %f\n", 1/2.0);
printf("1 / 2 = %f\n", 1.0/2.0);
```

# Relational Operators

```c++
 #include <stdio.h>

// == 
if (5 == 6)
    printf("5 is equal 6\n");
else
    printf("5 is not equal 6\n");
```

```c++
 #include <stdio.h>

// !=
if (5 != 6)
    printf("5 is not equal 6\n");
else
    printf("5 is equal 6\n");
```

```c++
 #include <stdio.h>

// >
if (5 > 6)
    printf("5 is greater than 6\n");
else
    printf("5 is not greater than 6\n");
```

```c++
 #include <stdio.h>

// <
if (5 < 6)
    printf("5 is less than 6\n");
else
    printf("5 is not less than 6\n");
```

```c++
 #include <stdio.h>

// >= 
if (5 >= 6)
    printf("5 is greater than or equal 6\n");
else
    printf("5 is not greater than or equal 6\n");
```

```c++
#include <stdio.h>

// <= 
if (5 <= 6)
    printf("5 is less than or equal 6\n");
else
    printf("5 is not less than or equal 6\n");
```

# Assignment Operators

```c++
#include <stdio.h>

// +=
int a = 5;
int b = 0;
int c = 0;

b += a;
c = c + a;

printf("b = %d\n", b);
printf("c = %d\n", c);
```

```c++
#include <stdio.h>

// -=
int a = 5;
int b = 0;
int c = 0;

b -= a;
c = c - a;

printf("b = %d\n", b);
printf("c = %d\n", c);
```

```c++
#include <stdio.h>

// *=
int a = 5;
int b = 0;
int c = 0;

b *= a;
c = c * a;

printf("b = %d\n", b);
printf("c = %d\n", c);
```

```c++
#include <stdio.h>

// /=
int a = 5;
float b = 1;
float c = 1;

b /= a;
c = c / a;

printf("b = %.1f\n", b);
printf("c = %.1f\n", c);
```

# Conditional Expression Operators

```c++
#include <stdio.h>

// ?:
int a = 5;
int b;  

b = ((a==5) ? 3 : 2 );

printf("b = %d", b); 
```

```c++
#include <stdio.h>

// ?:
int age = 18;

(age >= 18) 
? printf("eligible for voting") 
: printf("not eligible for voting"); 
```

# Function

```c++
#include <stdio.h>

int sum(int a, int b)
{
    return a+b;
}

printf("5 + 9 = %d", sum(5, 9));
```

```c++
#include <stdio.h>

void sum(int a, int b)
{
    printf("%d + %d = %d", a, b, a+b);
}

sum(5, 9);
```

```c++
#include <stdio.h>

void print_array(int arr[], int size)
{
    for (int i = 0; i < size; i++)
    {
        printf("%d\t", arr[i]);
    }
    puts("");
}

int a[] = {1, 2, 3, 4, 5};
int size = sizeof(a)/sizeof(a[0]);

print_array(a, size);
```

```c++
#include <stdio.h>

void print_array(int * arr, int size)
{
    for (int i = 0; i < size; i++)
    {
        printf("%d\t", arr[i]);
    }
    puts("");
}

int a[] = {1, 2, 3, 4, 5};
int size = sizeof(a)/sizeof(a[0]);

print_array(a, size);
```
