```c++
#include <stdio.h>

void print_with_space(int num)
{
    int temp = 0;
    int scale = 1;
    if (num < 0)
    {
        printf("-");
        num = -num;
    }

    while (num >= 1000)
    {
        temp = temp + scale * (num % 1000);
        num /= 1000;
        scale *= 1000;
    }

    printf("%d", num);

    while (scale != 1)
    {
        scale /= 1000;
        num = temp / scale;
        temp = temp % scale;
        printf(" %d", num);
    }

    puts("\n");
}
```

```c++
int i1 = -2039407152;
int i2 = 2039407151;

print_with_space(i2);
```
